## What is Pandas Profiling or ydata Profiling ?

Pandas Profiling automatically generates an exploratory data analysis (EDA) report from a pandas DataFrame.

## Installation : 

```python
!pip install ydata-profiling
```

### 1. Import libraries 

```python
from ydata_profiling import ProfileReport
import pandas as pd
```

### 2. Load your data

```python
df = pd.read_csv('ford.csv')
df.shape
```

### 3. Generate the profile

```python
profile = ProfileReport(df, title="Ford Report",explorative=True)


# Use explorative=True when:
# Dataset is small/medium
# You are in the early EDA phase
```

### 4. View the report

```python
profile.to_file("report.html")

# open report.html in browser to see the report
```
